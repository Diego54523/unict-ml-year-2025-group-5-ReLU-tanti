```python
# === Notebook cell: training end-to-end (usa il codice che hai già definito sopra) ===


# === FIX IMPORT PATH ===
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent  # ML-PROJECT/
sys.path.insert(0, str(PROJECT_ROOT))

print("Project root added to PYTHONPATH:", PROJECT_ROOT)

from pathlib import Path
import os
import torch
import src.models.radnet as radnet

if "__file__" not in globals():
    # prova a inferire la root progetto dalla cwd
    BASE_DIR = Path.cwd()
    # tenta alcune varianti comuni; prende la prima esistente
    candidates = [
        BASE_DIR / "archive" / "MRI",
        BASE_DIR.parent / "archive" / "MRI",
        BASE_DIR / ".." / "archive" / "MRI",
        BASE_DIR.parent.parent / "archive" / "MRI",
    ]
    for c in candidates:
        c = c.resolve()
        if c.is_dir():
            radnet.DATA_DIR = str(c)
            break

print("DATA_DIR =", radnet.DATA_DIR)
assert os.path.isdir(radnet.DATA_DIR), f"radnet.DATA_DIR non trovato: {radnet.DATA_DIR}"

try:
    _ = torch.cuda.is_available()
except Exception:
    pass

if os.name == "nt":
    radnet.NUM_WORKERS = 0
elif "COLAB_GPU" in os.environ or "KAGGLE_URL_BASE" in os.environ:
    radnet.NUM_WORKERS = min(radnet.NUM_WORKERS, 4)

print("NUM_WORKERS =", radnet.NUM_WORKERS)
print("CUDA available =", torch.cuda.is_available())

# 3) Lancia il training (chiama la tua main già definita sopra)
radnet.main()
```
