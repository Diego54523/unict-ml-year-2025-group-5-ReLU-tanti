As we can see the accuracy is, approximately, the same, so we concluded that the reason of such low accuracy lies in the choice of the ResNet18 as feature extractor.

The latter assuption brought us to try new paths. First we tried to train an our CNN for feature extration.

To train the CNN, after create an architecture inspired by the one of ResNet18, was trained with these parameters: lr = 0.001, epochs = 20 and, instead of using the Stochastic Gradient Descent, pointing to outweight it in fewer time, we used the Adaptive Moment Estimation (ADAM).
